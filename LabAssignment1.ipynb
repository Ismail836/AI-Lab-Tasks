# **Task 1**

```python
from collections import deque

# --- Graph Definition (Undirected, 0-Index, based on image_962100.png) ---
# Nodes: 0, 1, 2, 3, 4, 5
graph = {
    '0': ['1', '2'],        # Node 0 connected to 1, 2
    '1': ['0', '2'],        # Node 1 connected to 0, 2
    '2': ['0', '1', '3'], # Node 2 connected to 0, 1, 3
    '3': ['2', '4'],        # Node 3 connected to 2, 4
    '4': ['3', '5'],   # Node 4 connected to 3, 5
    '5': ['4']              # Node 5 connected to 4
}

# --- Depth-First Search (DFS) Implementation ---
def dfs(graph, start_node):
    stack = [start_node]
    visited = set()
    path = []

    while stack:
        vertex = stack.pop()

        if vertex not in visited:
            visited.add(vertex)
            path.append(vertex)

            # Use reverse sorted neighbors for consistent DFS path
            neighbors = sorted(graph.get(vertex, []), reverse=True)
            for neighbor in neighbors:
                if neighbor not in visited:
                    stack.append(neighbor)
    return path

# --- Breadth-First Search (BFS) Implementation ---
def bfs(graph, start_node):
    queue = deque([start_node])
    visited = {start_node}
    path = []

    while queue:
        vertex = queue.popleft()
        path.append(vertex)

        # Use sorted neighbors for consistent BFS path
        neighbors = sorted(graph.get(vertex, []))
        for neighbor in neighbors:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return path

# --- Execution ---
start = '0' # Start value is 0

print(f"--- Graph Search (0-Index, Starting at Node {start}) ---")
print("Corrected Graph Adjacency List (0-Index):")
for node, neighbors in graph.items():
    print(f"Node {node}: {', '.join(neighbors)}")

print("\n---------------------------------------------------")
# Run BFS
bfs_path = bfs(graph, start)
print(f"BFS Path: {' -> '.join(bfs_path)}")

# Run DFS
dfs_path = dfs(graph, start)
print(f"DFS Path: {' -> '.join(dfs_path)}")
print("---------------------------------------------------")
```

# **Task 4**

```python
import time
from collections import deque

# --- Problem Constants ---
MAX_A = 4    # Maximum capacity of Jug A (4 gallons)
MAX_B = 3    # Maximum capacity of Jug B (3 gallons)
GOAL = 2     # Target amount in Jug A
START_STATE = (0, 0)  # Starting state: (Jug A amount, Jug B amount)

def get_successors(a, b):
    """Generates all possible next states from the current state (a, b)."""
    successors = []

    # 1. Fill Jug A completely
    successors.append((MAX_A, b))

    # 2. Fill Jug B completely
    successors.append((a, MAX_B))

    # 3. Empty Jug A
    successors.append((0, b))

    # 4. Empty Jug B
    successors.append((a, 0))

    # 5. Pour A to B
    # Amount A can pour = minimum of (current A amount, space left in B)
    pour_a_to_b = min(a, MAX_B - b)
    successors.append((a - pour_a_to_b, b + pour_a_to_b))

    # 6. Pour B to A
    # Amount B can pour = minimum of (current B amount, space left in A)
    pour_b_to_a = min(b, MAX_A - a)
    successors.append((a + pour_b_to_a, b - pour_b_to_a))

    return [state for state in successors if 0 <= state[0] <= MAX_A and 0 <= state[1] <= MAX_B]


def solve_water_jug(search_type='BFS'):
    """Implements either BFS or DFS to find the solution."""
    start_time = time.perf_counter()

    # BFS uses a Queue (deque), DFS uses a Stack (list.pop())
    if search_type == 'BFS':
        # (current_state, path_to_current_state)
        data_structure = deque([(START_STATE, [START_STATE])])
    else: # DFS
        data_structure = [(START_STATE, [START_STATE])]

    visited = {START_STATE}
    steps_explored = 0

    # Memory usage is estimated by the maximum size of the data structure (queue/stack)
    # plus the visited set size at any point.
    max_memory_usage = len(data_structure) + len(visited)

    while data_structure:
        steps_explored += 1

        # Pop the state: from the left for BFS, from the right for DFS
        if search_type == 'BFS':
            current_state, path = data_structure.popleft()
        else:
            current_state, path = data_structure.pop()

        a, b = current_state

        # GOAL TEST: Check if Jug A has the required amount
        if a == GOAL:
            time_taken = time.perf_counter() - start_time
            return {
                'steps_taken': steps_explored,
                'time_taken': time_taken,
                'memory_usage': max_memory_usage,
                'path': path
            }

        # Generate and process successors
        for next_state in get_successors(a, b):
            if next_state not in visited:
                visited.add(next_state)
                new_path = path + [next_state]
                data_structure.append((next_state, new_path))

        # Update maximum memory usage estimate
        current_memory = len(data_structure) + len(visited)
        max_memory_usage = max(max_memory_usage, current_memory)

    # If the goal is not reachable
    time_taken = time.perf_counter() - start_time
    return {
        'steps_taken': steps_explored,
        'time_taken': time_taken,
        'memory_usage': max_memory_usage,
        'path': []
    }

# --- Execution and Comparison ---
print("--- TASK 5: Water Jug Problem (4-gal, 3-gal, Goal 2 in A) ---")

# Run BFS
bfs_result = solve_water_jug('BFS')

# Run DFS
dfs_result = solve_water_jug('DFS')

print("\n--- Performance Comparison ---")
print(f"{'Algorithm':<10} | {'Steps Explored':<16} | {'Time Taken (s)':<14} | {'Max Memory (States)':<20}")
print("-" * 65)
print(f"{'BFS':<10} | {bfs_result['steps_taken']:<16} | {bfs_result['time_taken']:<14.6f} | {bfs_result['memory_usage']:<20}")
print(f"{'DFS':<10} | {dfs_result['steps_taken']:<16} | {dfs_result['time_taken']:<14.6f} | {dfs_result['memory_usage']:<20}")


print("\n--- BFS Solution Path (Guaranteed Shortest) ---")
if bfs_result['path']:
    print(" -> ".join(map(str, bfs_result['path'])))
else:
    print("No solution found by BFS.")

print("\n--- DFS Solution Path ---")
if dfs_result['path']:
    print(" -> ".join(map(str, dfs_result['path'])))
else:
    print("No solution found by DFS.")
```

# **Task 5**

```python
import heapq

# --- Maze Definition ---
# The maze is represented as a 2D grid [Row][Column].
# R (Row) goes from 0 (bottom) to 5 (top).
# C (Column) goes from 0 (left) to 5 (right).
# 'W' represents a Wall (Blue/Obstacle). Other letters are Path Nodes.

# The grid data is indexed [Row][Column]
maze_grid = [
    # C: 0    1    2    3    4    5
    ['W', 'W', 'W', 'W', 'X', 'Y'], # R: 5
    ['R', 'S', 'T', 'U', 'W', 'V'], # R: 4
    ['M', 'N', 'W', 'O', 'P', 'Q'], # R: 3
    ['H', 'I', 'J', 'W', 'K', 'L'], # R: 2
    ['F', 'W', 'G', 'W', 'W', 'W'], # R: 1
    ['A', 'W', 'B', 'C', 'D', 'E']  # R: 0
]

MAZE_ROWS = len(maze_grid) # 6
MAZE_COLS = len(maze_grid[0]) # 6

# Helper to find coordinates (Row, Column) and the corresponding Node Label
def find_node_details(target):
    # # Helper function: Finds coordinates of Start ('A') and Goal ('Y') #
    for r in range(MAZE_ROWS):
        for c in range(MAZE_COLS):
            if maze_grid[r][c] == target:
                return (r, c), target
    return None, None

def get_label_from_coords(coord):
    # # Helper function: Gets the letter label (A, B, C, etc.) from coordinates #
    r, c = coord
    return maze_grid[r][c]

START_COORDS, _ = find_node_details('A')
GOAL_COORDS, GOAL_LABEL = find_node_details('Y')


# --- Heuristic Function: Manhattan Distance h(n) ---
def manhattan_distance(coord1, coord2):
    # # Heuristic: Calculates straight-line cost (Manhattan distance) to the goal #
    return abs(coord1[0] - coord2[0]) + abs(coord1[1] - coord2[1])


# --- A* Search Algorithm ---
def a_star_search(start, goal):
    # # A* Priority Queue: (f_cost, g_cost, coordinates, path_list) #
    h_start = manhattan_distance(start, goal)
    open_list = [(h_start, 0, start, [start])]

    # # g_costs: Stores the lowest actual cost (steps) to reach a node #
    g_costs = {start: 0}

    # # Movements: Up, Down, Left, Right #
    movements = [(-1, 0), (1, 0), (0, -1), (0, 1)] # (dr, dc)

    while open_list:
        # # Pop node with the lowest f_cost (highest priority) #
        f_cost, g_cost, current_coord, path = heapq.heappop(open_list)

        # # Goal Check #
        if current_coord == goal:
            return path

        # # Explore Neighbors #
        for dr, dc in movements:
            new_r, new_c = current_coord[0] + dr, current_coord[1] + dc
            neighbor_coord = (new_r, new_c)

            # # Check bounds and walls ('W') #
            if 0 <= new_r < MAZE_ROWS and 0 <= new_c < MAZE_COLS and maze_grid[new_r][new_c] != 'W':

                new_g_cost = g_cost + 1 # Cost to move one step is 1

                # # Check if this path to the neighbor is shorter #
                if neighbor_coord not in g_costs or new_g_cost < g_costs[neighbor_coord]:

                    g_costs[neighbor_coord] = new_g_cost
                    new_h_cost = manhattan_distance(neighbor_coord, goal)
                    new_f_cost = new_g_cost + new_h_cost

                    new_path = path + [neighbor_coord]

                    # # Add the new, improved path to the priority queue #
                    heapq.heappush(open_list, (new_f_cost, new_g_cost, neighbor_coord, new_path))

    return None # # No path found #

# --- Execution ---
print("--- TASK 6: A* Search Results (Alphabet Path) ---")

solution_path_coords = a_star_search(START_COORDS, GOAL_COORDS)

if solution_path_coords:
    # Convert coordinates to Node Labels (A, B, C, etc.)
    path_nodes = [get_label_from_coords(coord) for coord in solution_path_coords]

    print(f"\nPath found by A* Search (Total steps: {len(solution_path_coords) - 1}):")
    print(" -> ".join(path_nodes))

    # --- Visualization ---
    print("\nPath visualization on the grid (marked with '#'):")
    # # Create a copy of the maze grid for marking the path #
    path_map = [row[:] for row in maze_grid]

    for r, c in solution_path_coords:
        # # Mark path nodes with '#' except for Start ('A'), Goal ('Y'), and Walls ('W') #
        if path_map[r][c] not in ['A', 'Y', 'W']:
            path_map[r][c] = '#'

    # Print from Row 5 down to 0 for visualization clarity (matching the image)
    print("   0 1 2 3 4 5 (Column Index)")
    for r in range(MAZE_ROWS - 1, -1, -1):
        print(f"R{r}: {' '.join(path_map[r])}")
    print("(Row Index)")
else:
    print("\nNo path found from A to Y.")
```
